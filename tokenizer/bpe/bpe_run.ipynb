```python
from bpe import BPETokenizer
import pandas as pd
```

```python
text = open('../../data/1984.txt', 'r', encoding='utf-8').read()
print(len(text))
```

```python
tokenizer = BPETokenizer(pattern=None)
tokenizer.train(text, vocab_size=5000, verbose=True)
```

```python
new_vocab_range = range(257, len(tokenizer.vocab))
reverse_vocab = {value: key for key, value in tokenizer.vocab.items()}

for i in new_vocab_range:
    token = tokenizer.vocab.get(i)
    token_id = reverse_vocab.get(token)
    print(f'text to token id: {tokenizer.encode_single(token)}')
    print(f'token id to text: {tokenizer.decode_single(token_id)}')
```

```python
vocab_ids = list(tokenizer.vocab.keys())
vocab_ids = vocab_ids[257:]
```

```python
# encode the text based on the vocab to encode the book as token ids
# from the learned vocab
book_id_list = []

for t in text:
    book_id_list.append(tokenizer.encode(t))

print(book_id_list)
```

```python
for ctx, target in tokenizer.generate_io_pairs(book_id_list):
    #print(f'{ctx} ----> {target}')
    print(f'{tokenizer.decode(ctx)} ---> {tokenizer.decode(target)}')
```
